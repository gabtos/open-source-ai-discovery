```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

def export_license_table(
    df: pd.DataFrame,
    column: str,
    output_path: str,
    include_unknown: bool = True,
    sort_descending: bool = True,
    max_rows: int = None
):
    """
    Summarize unique values in a column (e.g., license distribution),
    compute proportions and cumulative percentages, and export as a PNG table.

    Parameters
    ----------
    df : pd.DataFrame
        Input DataFrame.
    column : str
        Name of the column to summarize (e.g., "license").
    output_path : str
        Filepath for the output PNG image.
    include_unknown : bool, default=True
        Whether to include missing (NaN) values as 'unknown'.
    sort_descending : bool, default=True
        Whether to sort by count descending.
    max_rows : int, optional
        Maximum number of rows to display and include in the PNG.
        If None, all rows are included.
    """

    # =========================================================
    # 1. Prepare the data
    # =========================================================
    if include_unknown:
        series = df[column].fillna('unknown')
    else:
        series = df[column].dropna()

    counts_df = series.value_counts(dropna=not include_unknown).reset_index()

    # Handle column naming (different pandas versions)
    if counts_df.columns.tolist() == ['index', column]:
        counts_df.columns = ['value', 'count']
    else:
        counts_df.columns = ['value', 'count']

    # =========================================================
    # 2. Ensure numeric type and calculate proportions
    # =========================================================
    counts_df['count'] = pd.to_numeric(counts_df['count'], errors='coerce')
    total = counts_df['count'].sum()

    counts_df['proportion_percent'] = (
        (counts_df['count'] / total) * 100
    ).round(2)

    counts_df['cumulative_percent'] = (
        (counts_df['count'].cumsum() / total) * 100
    ).round(2)

    # Ensure cumulative ends at exactly 100%
    if not counts_df.empty:
        counts_df.loc[counts_df.index[-1], 'cumulative_percent'] = 100.00

    # Sort if needed
    if sort_descending:
        counts_df = counts_df.sort_values(by='count', ascending=False).reset_index(drop=True)

    # Limit rows if requested
    if max_rows is not None and len(counts_df) > max_rows:
        counts_df = counts_df.head(max_rows)

    # =========================================================
    # 3. Style and display (for Jupyter)
    # =========================================================
    styled = (
        counts_df.style
        .background_gradient(subset=["count"], cmap="Blues")
        .format({
            "proportion_percent": "{:.2f}",
            "cumulative_percent": "{:.2f}"
        })
        .set_caption(f"{column.title()} Distribution (Top {len(counts_df)} Shown)")
    )

    display(styled)

    # =========================================================
    # 4. Export as PNG
    # =========================================================
    output_path = Path(output_path)
    output_path.parent.mkdir(parents=True, exist_ok=True)

    rows = len(counts_df)
    fig_height = max(2, 0.4 * rows)

    fig, ax = plt.subplots(figsize=(8, fig_height))
    ax.axis('off')

    table = ax.table(
        cellText=counts_df.values,
        colLabels=counts_df.columns,
        loc='center'
    )
    table.auto_set_font_size(False)
    table.set_fontsize(10)
    table.scale(1.2, 1.2)

    plt.savefig(output_path, bbox_inches='tight', dpi=300)
    plt.close()

    print(f"✅ Table saved to {output_path.resolve()} (rows displayed: {len(counts_df)})")

    return counts_df
```

```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_license_trends(
    df: pd.DataFrame,
    date_col: str = "date_released",
    license_col: str = "license",
    licenses_to_include: list = None,
    freq: str = "M",   # 'M' for month, 'Y' for year
    top_n: int = None,
    kind: str = "line",
    figsize=(12,6)
):
    """
    Plots how selected licenses are used over time.

    Parameters
    ----------
    df : pd.DataFrame
        Input DataFrame.
    date_col : str
        Name of the column with dates (e.g. 'created_at' or 'last_modified').
    license_col : str
        Name of the column with license names.
    licenses_to_include : list, optional
        A list of license names to include (e.g. ['mit', 'apache-2.0']).
        If None, all non-unknown licenses are used.
    freq : str, default='M'
        Time frequency for aggregation ('M' for month, 'Y' for year).
    top_n : int, optional
        If provided, only the top N most frequent licenses are plotted.
    kind : str, default='line'
        'line' or 'area' chart type.
    figsize : tuple, default=(12,6)
        Figure size for the plot.
    """

    # 1️⃣ Convert to datetime
    df = df.copy()
    df[date_col] = pd.to_datetime(df[date_col], errors="coerce")

    # 2️⃣ Clean up licenses
    df[license_col] = df[license_col].astype(str).str.strip().str.lower()
    df = df[df[license_col].notna() & (df[license_col] != "unknown")]

    # 3️⃣ Filter for licenses of interest
    if licenses_to_include:
        licenses = [l.lower() for l in licenses_to_include]
        df = df[df[license_col].isin(licenses)]

    # 4️⃣ Create period column (e.g., year-month)
    df["period"] = df[date_col].dt.to_period(freq).astype(str)

    # 5️⃣ Group and pivot
    grouped = (
        df.groupby(["period", license_col])
        .size()
        .reset_index(name="count")
    )

    pivoted = grouped.pivot(
        index="period", columns=license_col, values="count"
    ).fillna(0)

    # 6️⃣ Sort by date
    pivoted.index = pd.to_datetime(pivoted.index)
    pivoted = pivoted.sort_index()

    # 7️⃣ Optionally select top N
    if top_n:
        top_cols = pivoted.sum().sort_values(ascending=False).head(top_n).index
        pivoted = pivoted[top_cols]

    # 8️⃣ Plot
    plt.figure(figsize=figsize)

    if kind == "area":
        pivoted.plot.area(figsize=figsize, alpha=0.8)
    else:
        pivoted.plot(kind="line", linewidth=2, figsize=figsize)

    plt.title("License Usage Over Time")
    plt.xlabel("Date")
    plt.ylabel("Number of Models")
    plt.legend(title="License", bbox_to_anchor=(1.05, 1), loc="upper left")
    plt.tight_layout()
    plt.show()

    return pivoted

```

```python
import numpy as np
import pandas as pd

def parse_mixed_dates(s: pd.Series) -> pd.Series:
    def _parse_one(v):
        if pd.isna(v):
            return pd.NaT
        # normalize to string
        if isinstance(v, (int, float, np.integer, np.floating)):
            # decide seconds vs milliseconds by magnitude
            v = int(v)
            if v > 10**12:          # looks like microseconds
                return pd.to_datetime(v, unit="us", utc=True, errors="coerce")
            elif v > 10**11:        # milliseconds
                return pd.to_datetime(v, unit="ms", utc=True, errors="coerce")
            elif v > 10**9:         # seconds but very large (rare)
                return pd.to_datetime(v, unit="s", utc=True, errors="coerce")
            else:                   # likely seconds
                return pd.to_datetime(v, unit="s", utc=True, errors="coerce")
        # string branch
        v_str = str(v).strip()
        # empty or placeholder
        if v_str in {"", "null", "none", "0000-00-00", "NaT"}:
            return pd.NaT
        # try ISO first
        out = pd.to_datetime(v_str, utc=True, errors="coerce")
        if out is not pd.NaT:
            return out
        # try numeric-looking strings
        if v_str.isdigit():
            n = int(v_str)
            if len(v_str) >= 16:  # microseconds
                return pd.to_datetime(n, unit="us", utc=True, errors="coerce")
            elif len(v_str) >= 13:  # milliseconds
                return pd.to_datetime(n, unit="ms", utc=True, errors="coerce")
            else:  # seconds
                return pd.to_datetime(n, unit="s", utc=True, errors="coerce")
        return pd.NaT

    out = s.apply(_parse_one)

    # sanity filter: keep only plausible model-era dates
    lo = pd.Timestamp("2008-01-01", tz="UTC")  # HF & modern repos era
    hi = pd.Timestamp.utcnow() + pd.Timedelta(days=1)
    out = out.where((out >= lo) & (out <= hi))

    return out

# Use it
open_source_df["parsed_date"] = parse_mixed_dates(open_source_df["date_released"])
print("parsed min/max:", open_source_df["parsed_date"].min(), open_source_df["parsed_date"].max())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

def plot_license_trends(
    df: pd.DataFrame,
    date_col: str = "parsed_date",
    license_col: str = "license",
    licenses_to_include: list = None,
    top_n: int = None,
    freq: str = "M",      # "M" for monthly, "Y" for yearly
    kind: str = "line",   # "line" or "area"
    figsize=(12, 6)
):
    """
    Plots how licenses are used over time, with options to include
    only specific licenses or the top N most common ones.

    Parameters
    ----------
    df : pd.DataFrame
        Input DataFrame with license and date columns.
    date_col : str
        Column containing datetime objects (parsed already).
    license_col : str
        Column containing license names.
    licenses_to_include : list, optional
        A list of licenses to include (e.g., ['mit', 'apache-2.0']).
    top_n : int, optional
        If provided, automatically selects the top N most common licenses.
    freq : str, default='M'
        Time frequency for aggregation ('M' for month, 'Y' for year).
    kind : str, default='line'
        Plot type: 'line' or 'area'.
    figsize : tuple, default=(12,6)
        Figure size for the plot.

    Returns
    -------
    pd.DataFrame
        Pivoted DataFrame of license counts per time period.
    """

    # 1️⃣ Clean and prepare data
    df = df.copy()
    df = df[df[license_col].notna()]
    df[license_col] = df[license_col].astype(str).str.lower().str.strip()
    df = df[df[license_col] != "unknown"]

    if licenses_to_include:
        # use only specified licenses
        licenses = [l.lower() for l in licenses_to_include]
        df = df[df[license_col].isin(licenses)]

    if top_n:
        # automatically pick top N most common licenses
        top_licenses = df[license_col].value_counts().head(top_n).index
        df = df[df[license_col].isin(top_licenses)]

    # 2️⃣ Handle dates
    df[date_col] = pd.to_datetime(df[date_col], errors="coerce")
    df = df[df[date_col].notna()]
    df["period"] = df[date_col].dt.to_period(freq).dt.to_timestamp()

    # 3️⃣ Aggregate counts by period + license
    license_time = (
        df.groupby(["period", license_col])
          .size()
          .reset_index(name="count")
    )

    pivoted = (
        license_time.pivot(index="period", columns=license_col, values="count")
                    .fillna(0)
                    .sort_index()
    )

    # 4️⃣ Plot
    plt.figure(figsize=figsize)
    if kind == "area":
        pivoted.plot.area(figsize=figsize, alpha=0.8)
    else:
        pivoted.plot(kind="line", linewidth=2, figsize=figsize)

    plt.title("License Usage Over Time")
    plt.xlabel("Date")
    plt.ylabel("Number of Models")

    # Format x-axis for monthly labels
    ax = plt.gca()
    ax.xaxis.set_major_locator(mdates.MonthLocator(interval=1))
    ax.xaxis.set_major_formatter(mdates.DateFormatter("%b %Y"))
    plt.xticks(rotation=45, ha="right")

    plt.legend(title="License", bbox_to_anchor=(1.05, 1), loc="upper left")
    plt.tight_layout()
    plt.show()

    return pivoted
```

## Model Releases over time



```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

def plot_model_release_trends(
    df: pd.DataFrame,
    date_col: str = "parsed_date",
    freq: str = "M",          # 'M' for month, 'Y' for year, etc.
    cumulative: bool = False, # cumulative total over time
    start_date: str = None,
    end_date: str = None,
    figsize=(12, 6),
    month_label_fmt: str = "%b %Y"  # e.g. "%b %d %Y" for Jan 01 2024
):
    """
    Plot how many models were released over time (overall count per period).

    Parameters
    ----------
    df : pd.DataFrame
        Input DataFrame with a date column.
    date_col : str
        Column containing datetimes (or parseable strings).
    freq : str
        Pandas period frequency for aggregation.
    cumulative : bool
        If True, plot cumulative totals instead of per-period counts.
    start_date, end_date : str or None
        Optional inclusive date window (e.g., '2023-01-01', '2025-09-30').
    figsize : tuple
        Figure size in inches.
    month_label_fmt : str
        Formatter for x-axis date labels (e.g., '%b %Y' or '%b %d %Y').

    Returns
    -------
    pd.DataFrame
        Time series with counts (and cumulative if requested).
    """

    x = df.copy()

    # Robust datetime parsing; keep timezone if present, coerce invalid
    x[date_col] = pd.to_datetime(x[date_col], errors="coerce", utc=True)
    x = x[x[date_col].notna()]

    # Optional windowing
    if start_date is not None:
        start_ts = pd.Timestamp(start_date, tz="UTC")
        x = x[x[date_col] >= start_ts]
    if end_date is not None:
        end_ts = pd.Timestamp(end_date, tz="UTC")
        x = x[x[date_col] <= end_ts]

    # Aggregate by period
    ts = (
        x.assign(period=lambda d: d[date_col].dt.to_period(freq).dt.to_timestamp())
         .groupby("period")
         .size()
         .rename("models_released")
         .to_frame()
         .sort_index()
    )

    if cumulative:
        ts["models_released"] = ts["models_released"].cumsum()

    # Plot
    ax = ts.plot(
        kind="line",
        linewidth=2,
        figsize=figsize,
        title=("Cumulative " if cumulative else "") + "Models Released Over Time",
        legend=False
    )

    ax.set_xlabel("Date")
    ax.set_ylabel("Cumulative Models" if cumulative else "Number of Models")

    # Month-by-month ticks with your chosen format
    ax.xaxis.set_major_locator(mdates.MonthLocator(interval=1))
    ax.xaxis.set_major_formatter(mdates.DateFormatter(month_label_fmt))
    plt.xticks(rotation=45, ha="right")

    plt.tight_layout()
    plt.show()

    return ts
```

```python
# --- Config (edit these) ---
CSV_PATH   = "hf_models_by_author.csv"   # your existing CSV with model rows
OUT_DIR    = "authors/author_license_counts"    # where to save outputs
AUTHOR_COL = "owner"                              # column name for author/owner
LICENSE_COL= "license"                             # column name for license string
ID_COL     = "repo_id"                                  # optional: unique model id to drop dups (set None to skip)
# ---------------------------

import os, re
import pandas as pd
from pathlib import Path

Path(OUT_DIR).mkdir(parents=True, exist_ok=True)

# Load
df = pd.read_csv(CSV_PATH)

# Optional: drop duplicates by model id if your CSV may have repeats
if ID_COL and ID_COL in df.columns:
    df = df.drop_duplicates(subset=[ID_COL])

# Keep only needed columns; guard missing cols
missing = [c for c in [AUTHOR_COL, LICENSE_COL] if c not in df.columns]
if missing:
    raise ValueError(f"Missing required columns in CSV: {missing}")

work = df[[AUTHOR_COL, LICENSE_COL]].copy()

# Normalize author
work[AUTHOR_COL] = work[AUTHOR_COL].fillna("").astype(str).str.strip()
work.loc[work[AUTHOR_COL] == "", AUTHOR_COL] = "UNKNOWN_AUTHOR"

# Normalize and split license strings:
# - lower case
# - replace separators (comma/semicolon/slash/pipe) with commas
# - remove extra spaces
# - split into multiple rows (explode)
def normalize_license(s: str) -> str:
    s = (s or "").strip()
    if not s:
        return "unknown"
    s = s.lower()
    # common synonyms/variants
    synonyms = {
        "apache2": "apache-2.0",
        "apache 2.0": "apache-2.0",
        "apache-2": "apache-2.0",
        "mit license": "mit",
        "bsd-3": "bsd-3-clause",
        "bsd-3-clause license": "bsd-3-clause",
        "cc by 4.0": "cc-by-4.0",
        "cc-by": "cc-by-4.0",
        "cc-by v4": "cc-by-4.0",
        "cc-by-4": "cc-by-4.0",
        "cc-by 4.0": "cc-by-4.0",
        "creative commons attribution 4.0": "cc-by-4.0",
        "proprietary license": "proprietary",
        "unknown license": "unknown",
    }
    s = synonyms.get(s, s)
    return s

# Replace various separators with commas, then split
sep_pattern = re.compile(r"[;,/|]+")
work[LICENSE_COL] = (
    work[LICENSE_COL]
    .fillna("unknown")
    .astype(str)
    .str.replace(r"\s+", " ", regex=True)
    .str.strip()
    .str.replace(sep_pattern, ",", regex=True)
)

# Split and explode to one license per row
work = (
    work
    .assign(**{LICENSE_COL: work[LICENSE_COL].str.split(",")})
    .explode(LICENSE_COL, ignore_index=True)
)

# Final clean of license tokens
work[LICENSE_COL] = (
    work[LICENSE_COL]
    .astype(str)
    .str.strip()
    .pipe(lambda s: s.where(s != "", "unknown"))
    .map(normalize_license)
)

# TALL: counts per author x license
counts_tall = (
    work
    .groupby([AUTHOR_COL, LICENSE_COL], dropna=False)
    .size()
    .reset_index(name="count")
    .sort_values([AUTHOR_COL, "count"], ascending=[True, False])
)

# WIDE: pivot to one row per author with license columns
counts_wide = (
    counts_tall
    .pivot(index=AUTHOR_COL, columns=LICENSE_COL, values="count")
    .fillna(0)
    .astype(int)
    .sort_index()
)
counts_wide["TOTAL"] = counts_wide.sum(axis=1)
counts_wide = counts_wide.sort_values("TOTAL", ascending=False)

# Save
tall_path = Path(OUT_DIR) / "author_license_counts_tall.csv"
wide_path = Path(OUT_DIR) / "author_license_counts_wide.csv"
counts_tall.to_csv(tall_path, index=False)
counts_wide.to_csv(wide_path)

# Preview
print(f"Saved:\n  {tall_path}\n  {wide_path}")
display(counts_tall.head(20))
display(counts_wide.head(20))
```
